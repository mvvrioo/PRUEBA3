```python
dni = []
lista_nombre = []
edad = []
pais = []
ciudad = []
all = [dni,lista_nombre,edad,pais,ciudad]
salir1 = True
from itertools import cycle

def digito_verificador(dni):

  reversed_digits = map(int, reversed(str(dni)))

  factors = cycle(range(2, 8))

  s = sum(d * f for d, f in zip(reversed_digits, factors))

  return (-s) % 11

print("\t\t    Bienvenido")
print("\tRegistro Nacional De Las Personas")
print("\tRENAPER AutoServicio\n")

while True:
 system=int(input("Que Desea Hacer\n(0)Grabar\n(1)Buscar\n(2)Imprimir certificado\n(3)Eliminar persona ingresada\n(4)Salir\n-----------------------------------------\n"))


 if system==0:
   while True:
    dniusu=str(input("Ingrese su DNI:\n"))
    dniusu=dniusu.replace(".","").replace("-","")#este saca los puntos y guion o sirve para que metan el dni sin guiones o puntos
    dniusu1=dniusu[:-1] #esto me deja el dni sin el numero verificador
    numv=digito_verificador(dniusu1)#me calcula el numero verificador del dni dado
    dniveri=int(dniusu[-1])#borra todo menos el numero verificador que dio el usuario
    if dniveri=="k":
      dniveri=10
    if numv==dniveri:#compara el numero verificador que calculo la funcion con el que dio el usuario
      dni.append(dniusu)
      break
    else:
      print("INGRESE UN DNI VALIDO")

   while True:
    nombre=str(input("Ingrese su nombre y apellido\n"))


    if len(nombre)<8:
       print("INGRESE UN NOMBRE VALIDO")
    else:
       lista_nombre.append(nombre)
       break

   while True:
        edid=int(input("Ingrese su edad\n"))

        if edid<1:
         print("INGRESE EDAD VALIDA")
        else:
         edad.append(edid)
         break

   paisna=str(input("Ingrese su pais de nacimiento\n"))
   pais.append(paisna)

   ciudadn=str(input("Ingrese su ciudad de nacimiento\n"))
   ciudad.append(ciudadn)

 if system==1:
   salir1=False
   busqueda=str(input("Ingrese DNI a buscar(SIN PUNTOS NI GUION):\n"))
   for j in range(len(dni)):
     if busqueda==dni[j]:
      print(f"-----------------------------------\nDNI:{dni[j]}\nNOMBRE:{lista_nombre[j]}\nEDAD:{edad[j]}\nPAIS:{pais[j]}\nCIUDAD:{ciudad[j]}\n-----------------------------------\n")
      break
   else:
    print("------------------------------------------------------------")
    print("\tNo se ha encontrado a nadie con ese DNI")
    print("------------------------------------------------------------")

 if system==2:
   estado=int(input("(1)CERTIFICADO DE NACIMIENTO\n(2)CERTIFICADO CONYUGAL\n"))
   if estado==1:
     print("\tCERTIFICADO DE NACIMIENTO")
     dnina=int(input("Ingrese DNI de la persona(SIN PUNTOS NI GUION)\n"))
     nombrenaci=str(input("Ingrese NOMBRE de la persona\n"))
     paisnaci=str(input("Ingrese su pais:\n"))
     ciudadnaci=str(input("Ingrese su ciudad de nacimiento:\n"))
     nacimiento=int(input("Ingrese DIA DE NACIMIENTO\n"))
     nacimiento1=int(input("Ingrese MES DE NACIMIENTO\n"))
     nacimiento2=int(input("Ingrese AÑO DE NACIMIENTO\n"))
     print("-------------------------------------------------------------------------------")
     print("\tCERTIFICADO DE NACIMIENTO")
     print(f"DNI:{dnina}")
     print(f"NOMBRE:{nombrenaci}")
     print(f"NACIMIENTO:{nacimiento}-{nacimiento1}-{nacimiento2}")
     print(f"PAIS:{paisnaci}")
     print(f"CIUDAD:{ciudadnaci}")
     print("\t\tCERTIFICADO VALIDO POR REGISTRO NACIONAL DE LAS PERSONAS")
     print("-------------------------------------------------------------------------------")
   else:
     print("-------------------------------------------------------------------------------")
     print("\tCERTIFICADO ESTADO CONYUGAL")
     nombrecon=str(input("Ingrese su nombre:\n"))
     edadconyu=int(input("Ingrese su edad:\n"))
     dniconyu=str(input("Ingrese su DNI(SIN PUNTOS NI GUIONES):\n"))
     paisconyu=str(input("Ingrese su pais:\n"))
     ciudadconyu=str(input("Ingrese su ciudad de nacimiento:\n"))
     estadoconyu=str(input("Ingrese su ESTADO CONYUGAL:\n"))
     nacimientoe=int(input("Ingrese DIA DE NACIMIENTO \n"))
     nacimientoe1=int(input("Ingrese MES DE NACIMIENTO \n"))
     nacimientoe2=int(input("Ingrese AÑO DE NACIMIENTO \n"))
     print("\tCERTIFICADO ESTADO CONYUGAL")
     print(f"NOMBRE:{nombrecon}\nEDAD:{edadconyu}\nDNI:{dniconyu}\nPAIS:{paisconyu}\nCIUDAD:{ciudadconyu}\nESTADO CONYUGAL:{estadoconyu}\nNACIMIENTO:{nacimientoe}-{nacimientoe1}-{nacimientoe2}\n")
     print("\t\tCERTIFICADO VALIDO POR REGISTRO NACIONAL DE LAS PERSONAS")
     print("-------------------------------------------------------------------------------")






 if system==3:
   eliminar=str(input("Ingrese el DNI de la persona que desea eliminar(SIN PUNTOS NI GUION):\n"))
   for x in range(len(dni)):
     if dni[x]==eliminar:
      print("----------------------------------------------------------------------")
      print(f"\tSE HAN ELIMINADO TODOS LOS DATOS DE: {lista_nombre[x]} ")
      print("----------------------------------------------------------------------")
      del dni[x], pais[x],lista_nombre[x],edad[x],ciudad[x]
      break
     else:
      print("----------------------------------------------------------------------")
      print("No se ha encontrado ningun dato sobre ese DNI")
      print("----------------------------------------------------------------------")


 if system==4:
  salir=int(input("Desea salir?\n(1)Si\n(2)No\n "))
  if salir==1:
    print("-----------------------------------------------")
    print("\t\tUsted ha salido")
    print("\tMario Torres PROGRAM VERSION PY 2.7")
    print("-----------------------------------------------")
    break



```
